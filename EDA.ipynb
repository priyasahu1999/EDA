2. Which car brand (Company) has the highest variation in prices, and what does this tell us about the pricing
trends

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("Car Sale.csv")

# Check for null values
print(df.isnull().sum())

# Drop rows with nulls (if any)
df.dropna(inplace=True)

dealer_price_avg = df.groupby("Dealer_Name")["Price ($)"].mean().sort_values(ascending=False)
print(dealer_price_avg)

# Plot
dealer_price_avg.plot(kind='bar', figsize=(12, 5), title="Avg Car Price by Dealer")
plt.ylabel("Average Price ($)")
plt.show()
```

```python
sns.boxplot(x='Transmission', y='Price ($)', data=df)
plt.title("Price Distribution by Transmission Type")
plt.show()
```

```python
sns.boxplot(x='Dealer_Region', y='Price ($)', data=df)
plt.title("Price Distribution Across Regions")
plt.show()
```

6. How does the average selling price of cars vary by customer gender and annual income ?

```python
sns.boxplot(x='Gender', y='Price ($)', data=df)
plt.title("Price by Gender")
plt.show()

sns.scatterplot(x='Annual Income', y='Price ($)', hue='Gender', data=df)
plt.title("Price vs Income by Gender")
plt.show()
```

7.  What is the distribution of car prices by region, and how does the number of cars sold vary by region

```python
region_avg = df.groupby("Dealer_Region")["Price ($)"].mean()
region_count = df['Dealer_Region'].value_counts()

print(region_avg)
print(region_count)

# Dual Plot
fig, ax1 = plt.subplots()
region_avg.plot(kind='bar', ax=ax1, color='blue', position=0, width=0.4, label='Avg Price')
region_count.plot(kind='bar', ax=ax1, color='orange', position=1, width=0.4, label='Cars Sold', secondary_y=True)
plt.title("Avg Price and Cars Sold by Region")
plt.legend()
plt.show()
```

8.  How does the average car price differ between cars with different engine sizes


```python
df['Engine'] = df['Engine'].astype(str)  # In case it's not string

engine_price_avg = df.groupby('Engine')['Price ($)'].mean().sort_values(ascending=False)
print(engine_price_avg)

engine_price_avg.plot(kind='bar', title="Avg Price by Engine Type")
plt.ylabel("Price ($)")
plt.show()
```

9. How do car prices vary based on the customer’s annual income bracket?

```python
# Create income brackets
df['Income Bracket'] = pd.cut(df['Annual Income'], bins=[0, 30000, 60000, 100000, 150000, np.inf],
                               labels=["<30k", "30k-60k", "60k-100k", "100k-150k", "150k+"])

sns.boxplot(x='Income Bracket', y='Price ($)', data=df)
plt.title("Price vs Annual Income Bracket")
plt.show()
```

10. What are the top 5 car models with the highest number of sales, and how does their price distribution look

```python
top_models = df['Model'].value_counts().head(5).index.tolist()
top_models_df = df[df['Model'].isin(top_models)]

sns.boxplot(x='Model', y='Price ($)', data=top_models_df)
plt.title("Price Distribution of Top 5 Models")
plt.show()
```

11.  How does car price vary with engine size across different car colors, and which colors have the highest price

12.  Is there any seasonal trend in car sales based on the date of sale ?

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month

monthly_sales = df.groupby('Month')['Price ($)'].mean()
monthly_sales.plot(kind='line', marker='o', title="Monthly Avg Sale Price")
plt.xlabel("Month")
plt.ylabel("Avg Price")
plt.show()
```

6.  What are the most popular product keywords
    - Words like “wireless”, “bluetooth”, “headphones” indicate trending tech products.



```python
import re
from collections import Counter

keywords = df['product_name'].dropna().apply(lambda x: re.findall(r'\w+', x.lower()))
flat_keywords = [word for sublist in keywords for word in sublist if len(word) > 2]
popular_keywords = pd.Series(Counter(flat_keywords)).sort_values(ascending=False).head(10)
print(popular_keywords)
```
